```python
# Get the language column names from your DataFrame
languages_list = languages.columns[1:]  # Assumes the language columns start from the second column

# Create subplots for the line plots
fig, axes = plt.subplots(len(languages_list), 1, figsize=(10, len(languages_list) * 2), sharex=True)

# Create line plots for each language
for i, language in enumerate(languages_list):
    ax = axes[i]  # Get the corresponding axis for the subplot

    # Plot the line
    sns.lineplot(data=languages, x='Date', y=language, ax=ax)
    ax.set_title(language)  # Set the title for each subplot

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

Observation : Following languages were at their popularity peak in 2004 and became least popular in 2023
- Visual Basic
- PHP
- Perl
- Java
- Delphi/Pascal

```python
plt.figure(figsize=[10,6])
sns.lineplot(data=languages,x='Date',y='Visual Basic',label='Visual Basic')
sns.lineplot(data=languages,x='Date',y='Perl', label='Perl')
sns.lineplot(data=languages,x='Date',y='Delphi/Pascal', label='Pascal')
plt.xlabel('Date')
plt.ylabel('Language')
plt.title('Programming Languages whose popularity decreased continuously')
plt.show()
```
