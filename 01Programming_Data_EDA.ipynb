```python
# Convert the Date column to datetime
languages['Date'] = pd.to_datetime(languages['Date'])
languages.head()
```

```python
# Boxplot of Java popularity over time
sns.boxplot(y=languages['Java'])
plt.show()
sns.boxplot(y=languages['Python'])
plt.show()
sns.boxplot(y=languages['JavaScript'])
plt.show()
sns.boxplot(y=languages['C#'])
plt.show()
```

```python
sns.distplot(languages['Java'])
```

```python
# Removing the outliers
languages = languages[languages['Java'] < 25]
languages = languages[languages['Python'] < 25]
languages = languages[languages['JavaScript'] < 25]
languages = languages[languages['C#'] < 25]

# Boxplot of Java popularity over time
sns.boxplot(y=languages['Java'])
plt.show()
sns.boxplot(y=languages['Python'])
plt.show()
sns.boxplot(y=languages['JavaScript'])
plt.show()
sns.boxplot(y=languages['C#'])
plt.show()

```
