## How do I select a subset of a DataFrame?
### 1:How do I select specific columns from a DataFrame?

```python
# To select multiple columns, use a list of column names within the selection brackets [].
attack_name = df[['Attack','Name']]
attack_name
```

### How do I filter specific rows from a DataFrame?

```python
attack = df[df['Attack'] > 30]
attack
```

The condition inside the selection brackets titanic["Age"] > 35 checks for which rows the Age column has a value larger than 35:

```python
type(attack)
```

```python
# I am interested in rows which have attack 30 or 50 
attack_3050 = df[df['Attack'].isin([30,50])]
attack_3050
```

The above is equivalent to filtering by rows for which the class is either 2 or 3 and combining the two statements with an | (or) operator:


```python
attack_3050 = df[(df['Attack'] == 30) | (df['Attack'] == 50)]
attack_3050
```

When combining multiple conditional statements, each condition must be surrounded by parentheses (). Moreover, you can not use or/and but need to use the or operator | and the and operator &.

I want to work with data for which the Type2 is known.

```python
type2 = df[df['Type 2'].notna()]
type2
```

### How do I select specific rows and columns from a DataFrame?

I’m interested in the names of the pokemons having attack smaller than 35

```python
attack_35 = df.loc[df['Attack'] < 35 , 'Name']
attack_35
```

When using loc/iloc, the part before the comma is the rows you want, and the part after the comma is the columns you want to select.


```python
df.loc[df['Attack'] > 25, ['Name', 'Generation']]
```

I’m interested in rows 10 till 25 and columns 3 to 5.

```python
df.iloc[10:26,3:6]
```

When selecting specific rows and/or columns with loc or iloc, new values can be assigned to the selected data. For example, to assign the name anonymous to the first 3 elements of the third column:

```python
df.iloc[0:3,3] = 'anonymus'
```
